```python
import matplotlib.pyplot as plt
import numpy as np


group1 = [0.19970, 0.17438, 0.17285, 0.19970]
group2 = [0.20347, 0.16643, 0.16620, 0.15370, 0.14069]
group3 = [0.17110,	0.18574,	0.15694]
group4 = [0.18870,	0.18870,	0.18870]
group5 = [0.18211,	0.14793,	0.18374,	0.18374,	0.15248]

data = [group1, group2, group3, group4, group5]
x = [1, 2, 3, 4, 5]

fig, ax = plt.subplots(figsize=(8, 5))

# Boxplot with means shown
ax.boxplot(data, positions=x, widths=0.5, patch_artist=True, showmeans=True)

# Overlay mean ± std
for i, values in enumerate(data):
    values = np.array(values)
    mean = np.mean(values)
    std = np.std(values)
    ax.errorbar(x[i], mean, yerr=std, fmt='o', color='red', label='mean±std' if i == 0 else "")

# STANDARD ERROR:
# for i, values in enumerate(data):
#     values = np.array(values)
#     mean = np.mean(values)
#     se = np.std(values, ddof=1) / np.sqrt(len(values))
#     ax.errorbar(x[i], mean, yerr=se, fmt='o', color='blue', label='mean±SE' if i == 0 else "")


ax.set_xticks(x)
ax.set_xlabel('preprocessing level')
ax.set_ylabel('val/mAP')
ax.set_title('Max mAP scores by preprocessing level')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# t test

from scipy.stats import ttest_ind

t_stat, p_val = ttest_ind(group4, group5, equal_var=False)  # Welch's t-test recommended
print(f"t={t_stat:.4f}, p={p_val:.4f}")
```

```python
from scipy.stats import f_oneway

group1 = [0.19970, 0.17438, 0.17285, 0.19970]
group2 = [0.20347, 0.16643, 0.16620, 0.15370, 0.14069]
group3 = [0.17110,	0.18574,	0.15694]
group4 = [0.18870,	0.18870,	0.18870]
group5 = [0.18211,	0.14793,	0.18374,	0.18374,	0.15248]

f_stat, p_val = f_oneway(group1, group2, group3, group4, group5)
print(f"ANOVA F={f_stat:.4f}, p={p_val:.4f}")
```

None are statistically significant.
